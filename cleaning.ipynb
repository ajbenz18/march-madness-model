# clean 2022 data

```python
data2022 = pd.read_csv('2022full.csv')
data2022 = data2022[data2022['TEAM'] != 'TEAM']
data2022.reset_index(inplace=True, drop=True)

```

```python
# get seed and how far they got out of the df
seeds = []
posts = []
for index, row in data2022.iterrows():
    if index %2 == 0:
        continue
    vals = str(row['TEAM'])
    if vals == 'nan':
        seeds.append(np.nan)
        posts.append(np.nan)
    else:
        s = vals.split()[0].split(' ')[0]
        p = vals.split(',')[1][1:]
        if p[0] == '\\':
            p = p[3:]
        seeds.append(int(s))
        posts.append(p)
        
```

```python
data2022 = data2022[data2022['RK'].notna()]
data2022['SEED'] = seeds
data2022['POSTSEASON'] = posts
data2022.replace({'Finals':'2ND', 'CHAMPS':'Champions', 'Elite Eight':'E8', 'Final Four':'F4', 'Sweet Sixteen':'S16'}, inplace=True)
data2022['YEAR'] = [2022] * len(data2022)
```

# Clean 2021 data

```python
data2021 = pd.read_csv('2021full.csv')
data2021 = data2021[data2021['TEAM'] != 'TEAM']
data2021.reset_index(inplace=True, drop=True)

# get seed and how far they got out of the df
seeds = []
posts = []
for index, row in data2021.iterrows():
    if index %2 == 0:
        continue
    vals = str(row['TEAM'])
    if vals == 'nan':
        seeds.append(np.nan)
        posts.append(np.nan)
    else:
        s = vals.split()[0].split(' ')[0]
        p = vals.split(',')[1][1:]
        if p[0] == '\\':
            p = p[3:]
        seeds.append(int(s))
        posts.append(p)
        
data2021 = data2021[data2021['RK'].notna()]
data2021['SEED'] = seeds
data2021['POSTSEASON'] = posts
data2021.replace({'Finals':'2ND', 'CHAMPS':'Champions', 'Elite Eight':'E8', 'Final Four':'F4', 'Sweet Sixteen':'S16'}, inplace=True)
data2021['YEAR'] = [2021] * len(data2021)

data2021
```

# combine 2021 and 2022

```python
newdata = pd.concat([data2022, data2021], axis=0)
newdata
```

# Getting rid of formatting that is replacing records with dates

```python
records = list(newdata['REC'])
m2n = {'Jan':1, 'Feb':2, 'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}

for r in range(len(records)):
    val = records[r]
    values = val.split('-')
    second = values[1]
    if not second.isdigit():
        second = m2n[second]
        records[r] = str(second) + '-' + values[0]

newdata['REC'] = records

# Get number of wins from record
wins = []
for rec in newdata['REC']:
    ws = rec.split('-')[0]
    wins.append(int(ws))
    
newdata['W'] = wins
print(wins)
```

# Adding in 2024 data

Got this data in the same way I got the 2023 data, but I had to add in the seeding manually

```python
data2024 = pd.read_csv("trank_data_2024.csv", header=None)
pd.set_option('display.max_columns', None)

data2024.head()
```

```python
columns = {0: "TEAM", 1:"ADJOE", 2:"ADJDE", 3:"BARTHAG", 5:"W", 6:"G", 
          7: "EFG_O", 8: "EFG_D", 9:"FTR", 10:"FTRD", 11:"TOR", 12:"TORD",
          13:"ORB", 14:"DRB", 16:"2P_O", 17:"2P_D", 18:"3P_O", 19:"3P_D",
          15: "ADJ_T", 34:"WAB", 33:"SEED"}
data2024 = data2024.rename(columns=columns)

# drop columns we aren't using
to_drop = []
for c in data2024.columns:
    if c not in columns.values():
        to_drop.append(c)
data2024.drop(to_drop, axis=1, inplace=True)

data2024['YEAR'] = 2024
data2024['POSTSEASON'] = -1
data2024['CONF'] = np.nan
data2024['SEED'] = data2024['SEED'].fillna(-1)
data2024 = data2024[data2024["SEED"] > 0]
```
